```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqs

# Definir valores de R e C
R = 33e3  # 33 kΩ
C = 100e-9  # 100 nF

# Calcular a constante de tempo RC
tau = R * C

# Criar o sistema de função de transferência H(s) = 1 / (RCs + 1)
numerator = [1]
denominator = [tau, 1]

# Gerar frequências logarítmicas de 0.01 Hz a 100 kHz e converter para rad/s
frequencies_hz = np.logspace(-2, 5, num=100)  # Ajustar para 10 mHz a 100 kHz
w = 2 * np.pi * frequencies_hz

# Calcular resposta em frequência
w, h = freqs(numerator, denominator, worN=w)

# Calculando magnitude em dB e fase em graus
magnitude = 20 * np.log10(np.abs(h))
phase = np.degrees(np.angle(h))

# Plotar o diagrama de Bode para a magnitude
plt.figure(figsize=(10, 5))
plt.semilogx(frequencies_hz, magnitude, label='Magnitude', linewidth=3, color='blue')  # Linha mais grossa e azul
plt.title('Diagrama de Bode de Magnitude do Filtro')
plt.xlabel('Frequência [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid(which='both', axis='x', linestyle='--', linewidth=0.5)  # Apenas linhas verticais principais
plt.grid(which='major', axis='y', linestyle='--', linewidth=0.5)  # Apenas linhas horizontais principais
plt.minorticks_on()
plt.legend()
plt.show()

# Plotar o diagrama de Bode para a fase
plt.figure(figsize=(10, 5))
plt.semilogx(frequencies_hz, phase, label='Fase', linewidth=3, color='green')  # Linha mais grossa e verde
plt.title('Diagrama de Bode de Fase do Filtro')
plt.xlabel('Frequência [Hz]')
plt.ylabel('Fase [graus]')
plt.grid(which='both', axis='x', linestyle='--', linewidth=0.5)  # Apenas linhas verticais principais
plt.grid(which='major', axis='y', linestyle='--', linewidth=0.5)  # Apenas linhas horizontais principais
plt.minorticks_on()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqs

# Definir valores de R e C
R = 33e3  # 33 kΩ
C = 100e-9  # 100 nF
R_L = 22e3

K = R_L / (R + R_L)

# Calcular a constante de tempo RC
tau = R * C

# Criar o sistema de função de transferência H(s) = 1 / (RCs + 1/K)
numerator = [K]
denominator = [R * C, 1]

# Gerar frequências logarítmicas de 0.01 Hz a 100 kHz e converter para rad/s
frequencies_hz = np.logspace(-2, 5, num=100)  # Ajustar para 10 mHz a 100 kHz
w = 2 * np.pi * frequencies_hz

# Calcular resposta em frequência
w, h = freqs(numerator, denominator, worN=w)

# Calculando magnitude em dB e fase em graus
magnitude = 20 * np.log10(np.abs(h))
phase = np.degrees(np.angle(h))

# Plotar o diagrama de Bode para a magnitude
plt.figure(figsize=(10, 5))
plt.semilogx(frequencies_hz, magnitude, label='Magnitude', linewidth=3, color='blue')  # Linha mais grossa e azul
plt.title('Diagrama de Bode de Magnitude do Filtro')
plt.xlabel('Frequência [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid(which='both', axis='x', linestyle='--', linewidth=0.5)  # Apenas linhas verticais principais
plt.grid(which='major', axis='y', linestyle='--', linewidth=0.5)  # Apenas linhas horizontais principais
plt.minorticks_on()
plt.legend()
plt.show()

# Plotar o diagrama de Bode para a fase
plt.figure(figsize=(10, 5))
plt.semilogx(frequencies_hz, phase, label='Fase', linewidth=3, color='green')  # Linha mais grossa e verde
plt.title('Diagrama de Bode de Fase do Filtro')
plt.xlabel('Frequência [Hz]')
plt.ylabel('Fase [graus]')
plt.grid(which='both', axis='x', linestyle='--', linewidth=0.5)  # Apenas linhas verticais principais
plt.grid(which='major', axis='y', linestyle='--', linewidth=0.5)  # Apenas linhas horizontais principais
plt.minorticks_on()
plt.legend()
plt.show()
```
